```python
from numpy import dot
from numpy.linalg import norm

def get_cosine_simliarity(A, B):
    return dot(A, B) / (norm(A) * norm(B))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from kiwipiepy import Kiwi
from sklearn.metrics.pairwise import cosine_similarity

kiwi = Kiwi()
josa_tags = ["JKS","JKC","JKO","JKG","JKB","JKV","JKQ","JX","JC",
            "EC","EF","EP","ETN","ETM"]

def removes_josa(sentence):
    result = []
    for token in kiwi.tokenize(sentence):
        if token.tag in josa_tags:
            continue

        result.append(token.lemma)
    return " ".join(result)

result_df = list(map(removes_josa, result_df))

display(result_df)

count_v = CountVectorizer()
count_matrix = count_v.fit_transform(result_df)

cosine = cosine_similarity(count_matrix)
print(cosine.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1. user_retrieved_exp에서 원하는 컬럼만 선택
selected_cols = [4, 5, 6, 7, 14]

# 2. 각 행을 문자열로 합치기
user_retrieved_exp['combined_text'] = user_retrieved_exp[selected_cols].astype(str).agg(' '.join, axis=1)

# 3. 비교할 문자열 (예: 첫 번째 행 기준)
query_text = user_retrieved_exp['combined_text'].iloc[0]

# 4. result_df와 query_text를 합쳐서 벡터화
all_texts = result_df + [query_text]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_texts)

# 5. 코사인 유사도 계산 (query_text vs result_df 전체)
cosine_sim = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])  # 마지막(query_text) vs 앞(result_df)

# 6. 유사도 높은 15개 인덱스 추출
top15_indices = np.argsort(cosine_sim[0])[::-1][:15]

print("유사도 높은 15개 result_df 인덱스:")
print(top15_indices)

# 7. 실제 텍스트 출력
for idx in top15_indices:
    print(f"Index: {idx}, Text: {result_df[idx]}, Similarity: {cosine_sim[0][idx]:.4f}")
```
